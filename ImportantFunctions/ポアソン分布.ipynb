### 一時間に平均1.5人の人が来る際のポアソン分布の可視化

```python
x = range(7)

y = []
for i in range(len(x)):
    
    temp = (pow(1.5, x[i]) * np.exp(-1.5)) / math.factorial(x[i])
    y.append(temp)

z = np.polyfit(x, y, 6) #np.polyfit(x, y, n) : n 次式で 2 変数の回帰分析
f = np.poly1d(z)    

x_new = np.linspace(x[0], x[-1], 50)
y_new = f(x_new)
    
plt.plot(x, y, 'o', x_new, y_new)
```
